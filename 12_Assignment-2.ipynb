### Problem_1: Explain why we have to use the Exception class while creating a Custom Exception

  - When creating a custom exception in Python, it is generally recommended to derive the new exception class from the built-in Exception class or one of its subclasses. The Exception class serves as the base class for all exceptions in Python, providing a well-defined structure and behavior for handling and propagating exceptions.
  - This allows your custom exception to be compatible with the existing exception handling mechanisms in Python.
  - Python has a rich hierarchy of built-in exception classes, with Exception as the root. The inheritance structure allows for organizing and categorizing exceptions based on their characteristics

### Problem_3: What errors are defined in the ArithmeticError class? Explain any two with an example.

  - The ArithmeticError class in Python is a base class for all arithmetic-related exceptions. It represents errors that occur during mathematical operations. 
  - Some common errors defined in the `ArithmeticError` class include **FloatingPointError**, **OverflowError**, and **ZeroDivisionError**.

```python
## Example of FloatingPointError
try:
    result = 1.0 / 0.0
except ArithmeticError as e:
    print("Error: ",e)
```

```python
## Example of ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error: ",e)
```

### Problem_4: Why LookupError class is used? Explain with an example KeyError and IndexError.

  - The LookupError class in Python is a base class for exceptions that occur when a lookup or indexing operation fails. It represents errors that occur when accessing elements or values from data structures like lists, dictionaries, or sequences.
  - Two commonly encountered exceptions derived from the LookupError class are `KeyError` and `IndexError`.

```python
## Example of KeyError
dictionary= {"name":"Anmol","age":30}

try:
    value = dictionary["country"]
except KeyError:
    print("Error : key not found in dictionary")
```

```python
## Example of IndexError;
#-->This error occur when we try to access index that is out of range;
my_list=[1,2,3,4,5]

try:
    value=my_list[10]
except IndexError:
    print("Error: Index of range")
```

### Problem_5: Explain ImportError. What is ModuleNotFoundError?

  - `ImportError` is an exception that is raised when an import statement fails to find and load a module.
  - It occurs when there is an issue with importing a module or when the specified module cannot be found.
  - The ImportError exception is a base class for various import-related errors and can have subclasses depending on the specific import issue encountered.

```python
## ModuleNotFoundError occur when particular module not found or cannot be located
## It occurs when the specified module is not present in the Python environment or is not accessible in the current search paths.
try:
    import my_module
except ModuleNotFoundError:
    print("Error: Module not Found")
```
